```python
import pandas as pd
import sqlalchemy as db
import matplotlib.pyplot as plt
import seaborn as sns
import math



DATABASE_URL = "sqlite:///../db/database.db"
engine = db.create_engine(DATABASE_URL)

def load_data(engine):
    query = "SELECT * FROM houses"
    return pd.read_sql(query, con=engine)
```

```python
grupo = ["city", "area", "rooms", "bathroom", "parking_spaces", "floor", "animal", "furniture"]
```

```python
df = load_data(engine)
df.head()
```

```python
def plot_categoria_valor(df, categoria, valor, plot_type='bar'):
    
    plt.figure(figsize=(8,5))
    
    if plot_type == 'bar':
        sns.barplot(x='categoria', y='valor', data=df, ci=None)
    elif plot_type == 'box':
        sns.boxplot(x='categoria', y='valor', data=df)
    elif plot_type == 'scatter':
        sns.stripplot(x='categoria', y='valor', data=df, jitter=True)
    elif plot_type == 'violin':
        sns.violinplot(x='categoria', y='valor', data=df)
    else:
        raise ValueError("plot_type deve ser 'bar', 'box', 'scatter' ou 'violin'")
    
    plt.title(f'{plot_type.capitalize()} de valor por categoria')
    plt.show()
```
